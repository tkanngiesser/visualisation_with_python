```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
sns.set_style('white')
```

```python
"""
=============================
matplotlib plot distributions
=============================

Most important elements to plot one or more distributions (here chi2 with various df).

subplots
linestyles
xlim, ylim
labels
title
annotation
horizontal and vertical line
"""

x = np.linspace(0, 5, 1000)
fig,ax = plt.subplots(1,1)

linestyles = [':', '--', '-.', '-']
deg_of_freedom = [1, 2, 3, 4]
for df, ls in zip(deg_of_freedom, linestyles):
    ax.plot(x, stats.chi2.pdf(x, df), linestyle=ls, label=r'$df=%i$' % df)

plt.xlim(0, 5)
plt.ylim(0, 0.5)

plt.xlabel('$\chi^2$')
plt.ylabel(r'$f(\chi^2)$')
plt.title(r'$\chi^2\ \mathrm{Distribution}$')

ax.annotate('point offset from data',
            xy=(2, 0.22), xycoords='data',
            xytext=(10, 30), textcoords='offset points',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='left', verticalalignment='bottom')

ax.axvline(1.64, linestyle='--', color='black')
ax.axhline(0.40, linestyle='-.', color='black')

plt.legend()
plt.show()
```
